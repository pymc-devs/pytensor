# Overview of Pytensor

```python
import pytensor
import pytensor.tensor as pt
import matplotlib.pyplot as plt
```

```python
x = pt.tensor('x', shape=(None,))
sin_x = pt.sin(x)

sin_x.dprint()
```

```python
f = pytensor.function([x], sin_x)
```

```python
fig, ax = plt.subplots()
ax.plot(range(-10, 10), f(range(-10, 10)))
plt.show()
```
